## Choosing Search Algorithms

Out of the three uninformed searches (BFS, DFS, UCS), UCS seems to expand too much, and DFS does not guarantee optimality, so we only use BFS.

For Greedy Best First Search, the heuristic does not make too large of an impact on number of expanded nodes. Since Max Level and Set Level heuristics took the most time in Problem 2, we exclude these two heuristics.

Finally, A* search was amazingly slow with its computations. I excluded Max Level and Set Level heuristics which are complex heuristics and therefore take too long.

# Questions

## Which algorithm or algorithms would be most appropriate for planning in a very restricted domain (i.e., one that has only a few actions) and needs to operate in real time?

In restricted domains such as Problem 1, most algorithm have decent speed. However, since it is for real time operations, complicated heuristics should be excluded since they are slow to compute. Based on the threshold time, I would choose between Greedy BFS and A* search with Unmet Goals heuristic.

## Which algorithm or algorithms would be most appropriate for planning in very large domains (e.g., planning delivery routes for all UPS drivers in the U.S. on a given day)

BFS, DFS, UCS, and the A* search algorithms all have exponential number of nodes expanded for actions, so on large domains they will take a long time. Therefore, GBFS would be the best search algorithm. It does not guarantee optimality, but in practice it seems to find good enough plans. There isn't too much difference between heuristics.

## Which algorithm or algorithms would be most appropriate for planning problems where it is important to find only optimal plans?

For uninformed search, BFS and UCS are guaranteed to find the optimal plan, but not DFS. GBFS cannot guarantee an optimal plan since it is "greedy." A* search with an admissible heuristic would guarantee an optimal plan. The max level and set level heuristics are admissible, so those algorithms would be most appropriate.
